```python
'''
Implement the Image classification CNN model for classifying hand-written MNIST dataset by dividing the model into following 4 stages:
a.	Loading and preprocessing the image data
b.	Defining the model's architecture
c.	Training the model
d.	Estimating the model's performance
'''
```

```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt

import seaborn as sns
import tensorflow as tf
from tensorflow.keras.layers import BatchNormalization, Conv2D, Dense, Dropout, MaxPooling2D, Flatten
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import SGD
     
```

```python
# Load the training and testing data from CSV files
train = pd.read_csv("/home/tanisha/DL/LP-IV-datasets/MNIST(Ass2&3)/mnist_train.csv")
train
```

```python
x_train = train.drop(['label'], axis = 1).values
y_train = train['label'].values
x_train.shape
```

```python
x_train = x_train.reshape((60000, 28, 28, 1)).astype('float32')/255
x_train.shape
```

```python
test = pd.read_csv("/home/tanisha/DL/LP-IV-datasets/MNIST(Ass2&3)/mnist_test.csv")
test
     
```

```python

x_test = test.drop(['label'], axis = 1).values
y_test = test['label'].values
x_test = x_test.reshape((10000, 28, 28, 1)).astype('float32')/255

x_test.shape
```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), input_shape = (28, 28, 1), activation = 'relu'))
model.add(BatchNormalization())
model.add(Conv2D(64, (3,3),activation = 'relu'))
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(64, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation = 'softmax'))

sgd = SGD(learning_rate = 0.1)
model.compile(optimizer= sgd, loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
model.summary()
     
```

```python
r = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs = 3, batch_size = 64)
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
r.history.keys()
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f}')
```

```python

plt.plot(r.history['accuracy'], label = 'accuracy', color = 'green')
plt.plot(r.history['val_accuracy'], label = 'val_accuracy', color = 'red')
plt.legend()
```

```python
plt.plot(r.history['loss'], label = 'loss', color = 'red')
plt.plot(r.history['val_loss'], label = 'val_loss', color = 'blue')
plt.legend()
```
